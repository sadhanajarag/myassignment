## Q1. What is Bayes' theorem?

- Bayes' Theorem is a way of finding a probability when we know certain other probabilities.

The formula is:

                    P(A|B) =  P(A) * P(B|A)/P(B)
 


- How often A happens given that B happens, written P(A|B),
- How often B happens given that A happens, written P(B|A)
- and how likely A is on its own, written P(A)
- and how likely B is on its own, written P(B)

## Q2. What is the formula for Bayes' theorem?

                P(A|B) =  P(A)*P(B|A)/P(B)
                
 
- P(A|B) : How often A happens given that B already happened
- P(B|A) : How often B happens given that A already happened
- P(A):   how likely A is on its own, written P(A)
- P(B) : probability of B on its own P(B)

- Financial modeling - Bayes' theorem is used in finance to estimate the probability of different scenarios occurring and their potential impact on investments. It helps investors make informed decisions and manage their portfolios effectively.

- Machine learning - Bayes' theorem is used in machine learning algorithms to make predictions and classify data. It helps improve the accuracy of predictions and optimize the performance of machine learning models.

## Q4. What is the relationship between Bayes' theorem and conditional probability?

- Conditional Probability : The probability of occurrence of a certain event, say A, based on some other event whether B is true or not.

- The equation of conditional probability is:P(A|B)=P(A∩B)/P(B)

- Bayes Theorem:  includes two conditional probabilities for the events, say A and B.

- The equation of Bayes Theorem is:P(A|B)=P(B|A)×P(A)/P(B)

**Relation :** We know that the Bayes Theorem is derived from conditional probability, but they are further also related, like, if we know the conditional probability, we can use the Bayes rule to find out the reverse probabilities.
