## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

## Note: Also add title, xlabel, and ylabel to the plot.

Scatter Plot : 
            Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them. The scatter() method in the matplotlib library is used to draw a scatter plot. Scatter plots are widely used to represent relation among variables and how change in one affects the other. 

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("X Axis data")
plt.ylabel("Y Axis data")
plt.title("Scatter Plot")
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


What :
    
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. 
A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared,
while the other axis represents the measured values corresponding to those categories.

Why:
    
A bar chart is used when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data. 
From a bar chart, we can see which groups are highest or most common, and how other groups compare against the others.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.xlabel("Company")
plt.ylabel("Profit")
plt.bar(company,profit)
plt.title("Company profit Chart")
plt.show()
```

```python
plt.barh(company,profit)
plt.ylabel("Company")
plt.xlabel("Profit")
plt.title("Company profit Chart")
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


Box plot :
    
The box plot in matplotlib is mainly used to displays a summary of a set of data having properties 
like minimum, first quartile, median, third quartile, and maximum.

why:

 box plot which is also known as a whisker plot displays a summary of a set of data 
containing the minimum, first quartile, median, third quartile, and maximum. In a box 
plot, we draw a box from the first quartile to the third quartile. A vertical line goes through 
the box at the median. The whiskers go from each quartile to the minimum or maximum.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```
